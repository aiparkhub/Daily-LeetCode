## 解法方案
- 给定两个非空链表来表示两个非负整数, 且位数按照逆序方式存储, 它们的每个节点只存储单个数字, 将两数相加返回一个新的链表;

> 🧐 *Coding Ideas Type 1* 🧐

预先先构建一个空头节点, 然后尾节点从头节点开始向后不断生成新的节点, 遍历两个链表的公共部分, 每次相加相应位数字和进位, 分配到结果的链表中, 公共部分遍历完后再确定长的链表剩余的部分, 使用同样的方式再次进行遍历;

 - 需要注意的是遍历时每次都要更新进位, 不断计算和时有没有发生进位, 以防止之前数据的污染;
 - 对于python来说需要新的变量做游标来遍历两个链表, 不能直接用形参, 否则会修改原链表;
 - 注意最后可能的进位;
 

````python
# -*- coding:utf-8 -*-
#
# Geek International Park | 极客国际公园 & GeekParkHub | 极客实验室
# Website | https://www.geekparkhub.com
# Description | Open · Creation |
# Open Source Open Achievement Dream, GeekParkHub Co-construction has never been seen before.
#
# HackerParkHub | 黑客公园
# Website | https://www.hackerparkhub.org
# Description | In the spirit of fearless exploration, create unknown technology and worship of technology.
#
# AiParkHub | 人工智能公园
# Website | https://github.com/aiparkhub
# Description | Embark on the wave of AI and push the limits of machine intelligence.
#
# @GeekDeveloper : JEEP-711
# @Author : system
# @Version : 0.2.5
# @Program : 定义 两数相加 示例 | Definition Adding two numbers Example
# @File : topic_2_add_two_numbers.ipynb
# @Copyright © 2019 - 2020 AiParkHub-Organization. All rights reserved.


class ListNode:
    """
    Definition for singly-linked list
    """
    
    def __init__(self, x):
        """
        定义 初始化 方法 | Definition initialization method
        :param x: 
        """
        self.val = x
        self.next = None

class Solution:
    """
    定义 解答类 | Definition Solution Class
    """
    @staticmethod
    def add_two_numbers(l1, l2):
            """
            :type l1: ListNode
            :type l2: ListNode
            :rtype: ListNode
            """
            
            # 用p1, p2赋值为l1和l2, 以防止在操作改变原链表
            p1, p2 = l1, l2
            # 创建一个空链表用来返回结果, 分别有头节点和尾结点
            head = ListNode(0)
            tail = head
            # `carry`表示进位值
            carry = 0
            # 处理两个链表的公共部分, 也就是两个链表都不为空的部分
            while p1 and p2:
                # 计算当前位相加的和
                num = p1.val + p2.val + carry
                # 大于9应该向前进一位
                if num > 9:
                    num -= 10
                    carry = 1
                else:
                    carry = 0
                # 添加节点
                tail.next = ListNode(num)
                # 尾节点向后移动
                tail = tail.next
                # 移动两条链表的公共部分
                p1 = p1.next
                p2 = p2.next
            # 取长链表剩余的部分, 也就是未参与上面计算的部分
            if p2:
                # 如果p2较长,将p2剩余的部分赋值给p1, 只需要处理p1
                p1 = p2
            # 接下来处理长链表剩余分部分
            while p1:
                # 最近的一位要考虑一下是否有进位
                num = p1.val + carry
                if num > 9:
                    num -= 10
                    carry = 1
                else:
                    carry = 0
                # 添加节点
                tail.next = ListNode(num)
                tail = tail.next
                # 移动处理的链表
                p1 = p1.next
            # 如果最后仍然有进位需要再分配一个节点
            if carry:
                # 创建一个val为1的ListNode节点, 然后将tail向后移动一位
                tail.next = ListNode(1)
                tail = tail.next
            # 将所有的加和及进位都处理完毕, 现在将链表收尾
            tail.next = None
            # 将链表的头节点返回
            return head.next # 去除初始化为0的头节点


# 定义 主模块 | Definition Main module
if __name__ == "__main__":
    
    # 创建 对象实例 | Create object instance
    la = ListNode(2)
    la.next = ListNode(4)
    la.next.next = ListNode(3)
    
    # 创建 对象实例 | Create object instance
    lb = ListNode(5)
    lb.next = ListNode(6)
    lb.next.next = ListNode(4)
    
    # 调用方法 | Calling method
    ss = Solution.add_two_numbers(la, lb)
    
    # 输出打印结果 | Output print results
    print("val =>", ss.val)
    print("next.val =>", ss.next.val)
    print("next.next.val =>",ss.next.next.val)
    print("next.next.next =>", ss.next.next.next) 
````

因此这个地方可以考虑使用递归来求解, 具体可以分为以下情况:

 - 当两个链表均不为空时, 计算两个节点值与上一位进位的和sum, 取sum的个位数构建新的节点, 更新进位为su 的十位数, 令本节点的next指针指向下一位求和返回的节点;
 - 当两个链表其中一个为空时, 计算不为空的节点值与上一位进位的和sum, 更新进位为sum的十位数, 若进位不为0, 取sum的个位数构建新节点, 令本节点的 next 指针指向下一位求和返回的节点, 注意只传递不为空的链表；若进位为 0，则直接更新不为空节点的值为 sum，此时此链表之后的所有高位值都不会更新，因此返回此节点。
 - 若两个链表都为空, 判断进位是否为0, 若进位为0, 直接返回NULL, 否则构建值为进位值的新节点并返回此节点;
 
> 🧐 *Coding Ideas Type  2* 🧐

 - 跟plusOne, add Binary玩的同一个花样, 但是相对上个思路来说更加简单和简洁;
 - 使用递归调用简化算法

````python

# -*- coding:utf-8 -*-
#
# Geek International Park | 极客国际公园 & GeekParkHub | 极客实验室
# Website | https://www.geekparkhub.com
# Description | Open · Creation |
# Open Source Open Achievement Dream, GeekParkHub Co-construction has never been seen before.
#
# HackerParkHub | 黑客公园
# Website | https://www.hackerparkhub.org
# Description | In the spirit of fearless exploration, create unknown technology and worship of technology.
#
# AiParkHub | 人工智能公园
# Website | https://github.com/aiparkhub
# Description | Embark on the wave of AI and push the limits of machine intelligence.
#
# @GeekDeveloper : JEEP-711
# @Author : system
# @Version : 0.2.5
# @Program : 定义 两数相加 示例 | Definition Adding two numbers Example
# @File : topic_2_add_two_numbers.ipynb
# @Copyright © 2019 - 2020 AiParkHub-Organization. All rights reserved.


class ListNode:
    """
    Definition for singly-linked list
    """
    
    def __init__(self, x):
        """
        定义 初始化 方法 | Definition initialization method
        :param x: 
        """
        self.val = x
        self.next = None

class Solution(object):
    """
    定义 解答类 | Definition Solution Class
    """
    
    def add_two_numbers(self, l1, l2):
        """
        :type l1: ListNode
        :type l2: ListNode
        :rtype: ListNode
        """
        
        # 特殊情况
        if l1 is None :
            return l2
        if l2 is None :
            return l1
        # 如果相加小于10, 不需要进位
        if l1.val + l2.val < 10:
            l3 = ListNode(l1.val + l2.val)
            l3.next = self.add_two_numbers(l1.next, l2.next)
        # 相加大于等于10, 需要进位
        elif l1.val + l2.val >= 10:
            l3 = ListNode(l1.val + l2.val - 10)
            tmp = ListNode(1)
            tmp.next = None
            # 递归调用
            l3.next = self.add_two_numbers(l1.next, self.add_two_numbers(l2.next ,tmp))
        return l3


# 定义 主模块 | Definition Main module
if __name__ == "__main__":
    
    # 创建 对象实例 | Create object instance
    la = ListNode(2)
    la.next = ListNode(4)
    la.next.next = ListNode(3)
    
    # 创建 对象实例 | Create object instance
    lb = ListNode(5)
    lb.next = ListNode(6)
    lb.next.next = ListNode(4)
    
    # 创建 对象实例 | Create object instance
    s = Solution()
    
    # 调用方法 | Calling method
    ss = s.add_two_numbers(la, lb)
    
    # 输出打印结果 | Output print results
    print("val =>", ss.val)
    print("next.val =>", ss.next.val)
    print("next.next.val =>", ss.next.next.val)
    print("next.next.next=>", ss.next.next.next)
````
