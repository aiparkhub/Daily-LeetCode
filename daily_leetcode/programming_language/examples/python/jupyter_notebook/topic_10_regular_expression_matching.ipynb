````python
# -*- coding:utf-8 -*-
#
# Geek International Park | æå®¢å›½é™…å…¬å›­ & GeekParkHub | æå®¢å®éªŒå®¤
# Website | https://www.geekparkhub.com
# Description | Open Â· Creation |
# Open Source Open Achievement Dream, GeekParkHub Co-construction has never been seen before.
#
# HackerParkHub | é»‘å®¢å…¬å›­
# Website | https://www.hackerparkhub.org
# Description | In the spirit of fearless exploration, create unknown technology and worship of technology.
#
# AiParkHub | äººå·¥æ™ºèƒ½å…¬å›­
# Website | https://github.com/aiparkhub
# Description | Embark on the wave of AI and push the limits of machine intelligence.
#
# @GeekDeveloper : JEEP-711
# @Author : system
# @Version : 0.2.5
# @Program : å®šä¹‰ æ­£åˆ™è¡¨è¾¾å¼åŒ¹é… ç¤ºä¾‹ | Definition Regular Expression Matching Example
# @File : topic_10_regular_expression_matching.ipynb
# @Copyright Â© 2019 - 2020 AiParkHub-Organization. All rights reserved.


class Solution(object):
    """
    å®šä¹‰ è§£ç­”ç±» | Definition Solution Class
    """
    
    def isMatch(self, s, p):
        """
        å®šä¹‰ åŒ¹é… æ–¹æ³• | Define matching method
        :type s: str
        :type p: str
        :rtype: bool
        """
        
        def helper(s, i, p, j):
            """
            å®šä¹‰ è¾…åŠ© æ–¹æ³• | Define helper methods
            :param s: 
            :param i: 
            :param p: 
            :param j: 
            :return: 
            """
            
            if j == -1:
                return i == -1
            if i == -1:
                if p[j] != '*':
                    return False
                return helper(s, i, p, j-2)
            if p[j] == '*':
                if p[j-1] == '.' or p[j-1] == s[i]:
                    if helper(s, i-1, p, j):
                        return True
                return helper(s, i, p, j-2)
            if p[j] == '.' or p[j] == s[i]:
                return helper(s, i-1, p, j-1)
            return False
        return helper(s, len(s)-1, p, len(p)-1)


# å®šä¹‰ ä¸»æ¨¡å— | Definition Main module
if __name__ == '__main__':
    # å®šä¹‰ å˜é‡ | Definition variable
    s = 'abc'
    p = 'a*abc'
    
    # åˆ›å»º å¯¹è±¡å®ä¾‹ | Create object instance
    ss = Solution()
    
    # è°ƒç”¨æ–¹æ³• è¾“å‡ºæ‰“å°ç»“æœ | Call method Output print result
    print(ss.isMatch(s, p))
        
````

````python

# -*- coding:utf-8 -*-
#
# Geek International Park | æå®¢å›½é™…å…¬å›­ & GeekParkHub | æå®¢å®éªŒå®¤
# Website | https://www.geekparkhub.com
# Description | Open Â· Creation |
# Open Source Open Achievement Dream, GeekParkHub Co-construction has never been seen before.
#
# HackerParkHub | é»‘å®¢å…¬å›­
# Website | https://www.hackerparkhub.org
# Description | In the spirit of fearless exploration, create unknown technology and worship of technology.
#
# AiParkHub | äººå·¥æ™ºèƒ½å…¬å›­
# Website | https://github.com/aiparkhub
# Description | Embark on the wave of AI and push the limits of machine intelligence.
#
# @GeekDeveloper : JEEP-711
# @Author : system
# @Version : 0.2.5
# @Program : å®šä¹‰ æ­£åˆ™è¡¨è¾¾å¼åŒ¹é… ç¤ºä¾‹ | Definition Regular Expression Matching Example
# @File : topic_10_regular_expression_matching.ipynb
# @Copyright Â© 2019 - 2020 AiParkHub-Organization. All rights reserved.


class Solution(object):
    """
    å®šä¹‰ è§£ç­”ç±» | Definition Solution Class
    """
    
    def isMatch(self, s, p):
        """
        å®šä¹‰ åŒ¹é… æ–¹æ³• | Define matching method
        :type s: str
        :type p: str
        :rtype: bool
        """
        
        m, n = len(s), len(p)
        dp = [ [0 for i in range(n+1)] for j in range(m+1)]

        dp[0][0] = 1

        # init the first line
        for i in range(2,n+1):
            if p[i-1] == '*':
                dp[0][i] = dp[0][i-2]

        for i in range(1,m+1):
            for j in range(1,n+1):
                if p[j-1] == '*':
                    if p[j-2] != s[i-1] and p[j-2] != '.':
                        dp[i][j] = dp[i][j-2]
                    elif p[j-2] == s[i-1] or p[j-2] == '.':
                        dp[i][j] = dp[i-1][j] or dp[i][j-2]

                elif s[i-1] == p[j-1] or p[j-1] == '.':
                    dp[i][j] = dp[i-1][j-1]
        return dp[m][n] == 1 


# å®šä¹‰ ä¸»æ¨¡å— | Definition Main module
if __name__ == '__main__':
    # å®šä¹‰ å˜é‡ | Definition variable
    s = 'abc'
    p = 'a*abc'
    
    # åˆ›å»º å¯¹è±¡å®ä¾‹ | Create object instance
    ss = Solution()
    
    # è°ƒç”¨æ–¹æ³• è¾“å‡ºæ‰“å°ç»“æœ | Call method Output print result
    print(ss.isMatch(s, p))
````

   - **s[i] != p[j-1] && p[j-1] != '.'** ï¼Œé‚£ä¹ˆ dp[i][j] = dp[i][j-2] ï¼Œä¹Ÿå°±æ˜¯ * å·å‰é¢çš„é‚£ä¸ªå­—ç¬¦åœ¨åŒ¹é…çš„è¿‡ç¨‹å½“ä¸­ä¸€ä¸ªéƒ½ä¸ä½¿ç”¨;
   - **elseï¼Œé‚£ä¹ˆdp[i][j] = dp[i-1][j] || dp[i][j-1] || dp[i][j-2]**, ä¹Ÿå°±æ˜¯è¯´è¦ä¹ˆä½¿ç”¨ '*' å·è¿›è¡ŒåŒ¹é…( dp[i-1][j] ), è¦ä¹ˆåªä½¿ç”¨ '*' å·å‰é¢çš„é‚£ä¸ªå­—ç¬¦åŒ¹é…ï¼Œä¸ä½¿ç”¨ '*' åŒ¹é…( dp[i][j-1] ), è¦ä¹ˆ '*' å·å‰é¢çš„é‚£ä¸ªå­—ç¬¦åœ¨åŒ¹é…çš„è¿‡ç¨‹å½“ä¸­ä¸€ä¸ªéƒ½ä¸ä½¿ç”¨ï¼ˆ dp[i][j-2] ï¼‰,åªè¦æœ‰ä¸€ä¸ªæ˜¯ true åˆ™èƒ½å¤ŸåŒ¹é…;
   
æœ€åè€ƒè™‘ä¸€ä¸‹è¾¹ç•Œæ¡ä»¶, ä¸€å¼€å§‹ i æ˜¯ä» 1 åˆ° len(s) ï¼Œj æ˜¯ 1 åˆ° len(p) ã€‚é¦–å…ˆä¸€æ¥å°±èƒ½æƒ³åˆ° dp[0][0] æ˜¯trueï¼Œå…¶ä»–éƒ½æ˜¯ falseã€‚ä½†æ˜¯è¿™ä¸ªè¾¹ç•Œæ¡ä»¶æ˜¯ä¸å¤Ÿçš„ã€‚æ¯”å¦‚ isMatch("aab", "c * a * b") ï¼Œdp[1][3] åº”è¯¥æ˜¯ä» dp[0][2] è½¬ç§»è¿‡æ¥çš„, æ‰€ä»¥éœ€è¦æ›´å¤šçš„è¾¹ç•Œæ¡ä»¶ï¼Œä¹Ÿå°±æ˜¯ä¸€å¼€å§‹çš„ * æ˜¯èƒ½åŒ¹é…ç©ºå­—ç¬¦ä¸²çš„ã€‚æ‰€ä»¥æˆ‘æŠŠ i æ”¹åˆ°äº†ä» 0 å¼€å§‹,å¹¶ä¸”ç¬¬äºŒæ¡ä¹Ÿæ·»åŠ äº† i=0ï¼Œdp[i][j] = dp[i][j-2]

````python
# -*- coding:utf-8 -*-
#
# Geek International Park | æå®¢å›½é™…å…¬å›­ & GeekParkHub | æå®¢å®éªŒå®¤
# Website | https://www.geekparkhub.com
# Description | Open Â· Creation |
# Open Source Open Achievement Dream, GeekParkHub Co-construction has never been seen before.
#
# HackerParkHub | é»‘å®¢å…¬å›­
# Website | https://www.hackerparkhub.org
# Description | In the spirit of fearless exploration, create unknown technology and worship of technology.
#
# AiParkHub | äººå·¥æ™ºèƒ½å…¬å›­
# Website | https://github.com/aiparkhub
# Description | Embark on the wave of AI and push the limits of machine intelligence.
#
# @GeekDeveloper : JEEP-711
# @Author : system
# @Version : 0.2.5
# @Program : å®šä¹‰ æ­£åˆ™è¡¨è¾¾å¼åŒ¹é… ç¤ºä¾‹ | Definition Regular Expression Matching Example
# @File : topic_10_regular_expression_matching.ipynb
# @Copyright Â© 2019 - 2020 AiParkHub-Organization. All rights reserved.


class Solution:
    """
    å®šä¹‰ è§£ç­”ç±» | Definition Solution Class
    """
    
    def isMatch(self, s, p):
        """
        å®šä¹‰ åŒ¹é… æ–¹æ³• | Define matching method
        :type s: str
        :type p: str
        :rtype: bool
        """
        
        dp = []
        dp = [[False for i in range(len(p)+1)] for j in range(len(s)+1)]
        dp[0][0] = True
        # print(dp)
        for i in range(len(s)+1):
            for j in range(1, len(p)+1):
                if i > 0 and (s[i-1] == p[j-1] or p[j-1] == '.'):
                    # print("1111111",i)
                    # print("2222222",j)
                    dp[i][j] = dp[i-1][j-1]
                    # print("3333333",dp[i-1][j-1])
                if p[j-1] == '*':
                    if i == 0 or (s[i-1] != p[j-2] and p[j-2] != '.'):
                        # print("4444444",i)
                        # print("5555555",j)
                        dp[i][j] = dp[i][j-2]
                        # print("6666666",dp[i][j-2])
                    else:
                        dp[i][j] = dp[i-1][j] or dp[i][j-1] or dp[i][j-2]
        return dp[len(s)][len(p)]
    

# å®šä¹‰ ä¸»æ¨¡å— | Definition Main module
if __name__ == '__main__':
    # å®šä¹‰ å˜é‡ | Definition variable
    s = 'aa'
    p = 'a*'
    
    # åˆ›å»º å¯¹è±¡å®ä¾‹ | Create object instance
    ss = Solution()
    
    # è°ƒç”¨æ–¹æ³• è¾“å‡ºæ‰“å°ç»“æœ | Call method Output print result
    print(ss.isMatch(s, p))
````
