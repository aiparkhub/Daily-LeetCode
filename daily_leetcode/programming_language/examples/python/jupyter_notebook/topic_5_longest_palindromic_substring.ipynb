````python
# -*- coding:utf-8 -*-
#
# Geek International Park | æå®¢å›½é™…å…¬å›­ & GeekParkHub | æå®¢å®éªŒå®¤
# Website | https://www.geekparkhub.com
# Description | Open Â· Creation |
# Open Source Open Achievement Dream, GeekParkHub Co-construction has never been seen before.
#
# HackerParkHub | é»‘å®¢å…¬å›­
# Website | https://www.hackerparkhub.org
# Description | In the spirit of fearless exploration, create unknown technology and worship of technology.
#
# AiParkHub | äººå·¥æ™ºèƒ½å…¬å›­
# Website | https://github.com/aiparkhub
# Description | Embark on the wave of AI and push the limits of machine intelligence.
#
# @GeekDeveloper : JEEP-711
# @Author : system
# @Version : 0.2.5
# @Program : å®šä¹‰ æœ€é•¿å›æ–‡å­ä¸² ç¤ºä¾‹ | Definition Longest palindrome substring Example
# @File : topic_5_longest_palindromic_substring.ipynb
# @Copyright Â© 2019 - 2020 AiParkHub-Organization. All rights reserved.


class Solution3(object):
    """
    å®šä¹‰ è§£ç­”ç±» | Definition Solution Class
    """
    
    def longestPalindrome(self, s):
        """
        å®šä¹‰ æœ€é•¿å›æ–‡ æ–¹æ³• | Definition longest palindrome method
        :type s: str
        :rtype: str
        """
        
        def preProcess(s):
            """
            å®šä¹‰ å¤„ç† æ–¹æ³• | Definition process method
            :param s: 
            :return: 
            """
            if not s:
                return ['^', '&']
            T = ['^']
            for i in s:
                T += ['#', i]
            T += ['#', '$']
            return T
        
        T = preProcess(s)
        P = [0] * len(T)
        id, mx = 0, 0
        
        for i in range(1, len(T)-1):
            j = 2 * id - i
            if mx > i:
                P[i] = min(P[j], mx-i)
            else:
                P[i]= 0
            while T[i+P[i]+1] == T[i-P[i]-1]:
                P[i] += 1
            if i + P[i] > mx:
                id, mx = i, i + P[i]
        # ä¿å­˜çš„æ˜¯å½“å‰æœ€å¤§å›æ–‡å­ä¸²ä¸­å¿ƒä½ç½®çš„index
        max_i = P.index(max(P))
        start = int((max_i - P[max_i] - 1) / 2)
        res = s[start: start + P[max_i]]
        return res
 
 
if __name__ == '__main__':
    # å®šä¹‰ å˜é‡ | Definition variable
    nums = "acbcd"
    
    # åˆ›å»º å¯¹è±¡å®ä¾‹ | Create object instance
    s_3 = Solution3()
    
    # è°ƒç”¨æ–¹æ³• è¾“å‡ºæ‰“å°ç»“æœ | Call method Output print result
    print("res_4 =>", s_3.longestPalindrome(nums))
````
