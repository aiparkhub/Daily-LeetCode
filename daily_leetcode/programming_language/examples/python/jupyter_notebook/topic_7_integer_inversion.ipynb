</Description>

## 编码难度：(简单 | Simple) 🌟

## 解法方案
> 🧐 *Coding Ideas Type 1* 🧐

- x > 0, flag = 1; x < 0, flag = -1
- 将x转成字符串 s = str(abs(x))，然后再反转字符串 s1 = s[::-1]
- 字符串再转为整数：x1 = int(s1) * flag
- 判断是否溢出, 将数据返回



````python
# -*- coding:utf-8 -*-
#
# Geek International Park | 极客国际公园 & GeekParkHub | 极客实验室
# Website | https://www.geekparkhub.com
# Description | Open · Creation |
# Open Source Open Achievement Dream, GeekParkHub Co-construction has never been seen before.
#
# HackerParkHub | 黑客公园
# Website | https://www.hackerparkhub.org
# Description | In the spirit of fearless exploration, create unknown technology and worship of technology.
#
# AiParkHub | 人工智能公园
# Website | https://github.com/aiparkhub
# Description | Embark on the wave of AI and push the limits of machine intelligence.
#
# @GeekDeveloper : JEEP-711
# @Author : system
# @Version : 0.2.5
# @Program : 定义 整数反转 示例 | Definition Integer inversion Example
# @File : topic_7_integer_inversion.ipynb
# @Copyright © 2019 - 2020 AiParkHub-Organization. All rights reserved.


class Solution:
    """
    定义 解答类 | Definition Solution Class
    """
    
    def reverse(self, x):
        """
        定义 反转 方法 | Definition inversion method
        :type x: int
        :rtype: int
        """
        # 如果是负数需要注意转化为绝对值
        flag = 1
        x_1 = 0
        if x < 0:
            flag = -1
            x = int(str(abs(x))[::-1])
            x_1 = x * flag
        else:
            flag = 1
            x = int(str(x)[::-1])
            x_1 = x * flag
        if x_1 > 2**31-1 or x_1 < -2**31:
            return 0
        else:
            return x_1


# 定义 主模块 | Definition Main module
if __name__ == '__main__':
    
    # 定义 变量 | Definition variable
    x = 120
    y = -456
    
    # 创建 对象实例 | Create object instance
    s = Solution()
    
    # 调用方法 输出打印结果 | Call method Output print result
    print("x =>", s.reverse(x))
    print("y =>", s.reverse(y))
    
````

> 🧐 *Coding Ideas Type 2* 🧐

 - 按照低位摘除, 在按高位加上去其实就可以了;
 - 利用循环, 每次将最后一位取出来(求余), 然后将原来的数据除以10, 取整就可以每次都得到个位数;
 - 按照取第二步, 反向操作取出来一个最后的个位数, 然后乘以10, 再加上每一次得到的个位数就实现了反转;
 - 注意python中对一个数求余时, 结果与求余的符号相同, 比如 5 % 2 = 1, 5 % (-2) = -1


````python
# -*- coding:utf-8 -*-
#
# Geek International Park | 极客国际公园 & GeekParkHub | 极客实验室
# Website | https://www.geekparkhub.com
# Description | Open · Creation |
# Open Source Open Achievement Dream, GeekParkHub Co-construction has never been seen before.
#
# HackerParkHub | 黑客公园
# Website | https://www.hackerparkhub.org
# Description | In the spirit of fearless exploration, create unknown technology and worship of technology.
#
# AiParkHub | 人工智能公园
# Website | https://github.com/aiparkhub
# Description | Embark on the wave of AI and push the limits of machine intelligence.
#
# @GeekDeveloper : JEEP-711
# @Author : system
# @Version : 0.2.5
# @Program : 定义 整数反转 示例 | Definition Integer inversion Example
# @File : topic_7_integer_inversion.ipynb
# @Copyright © 2019 - 2020 AiParkHub-Organization. All rights reserved.


class Solution:
    """
    定义 解答类 | Definition Solution Class
    """ 
    
    def reverse(self, x):
        """
        定义 反转 方法 | Definition inversion method
        :type x: int
        :rtype: int
        """
        num = 0
        flag = 1
        if x > 0:
            flag = 1
        else:
            flag = -1
        while x != 0:
            num = num * 10 + x % (10 * flag)
            x = int(x / 10)
        if num > 2**31-1 or num < -2**31:
            return 0
        else:
            return num


# 定义 主模块 | Definition Main module        
if __name__ == '__main__':
    
    # 定义 变量 | Definition variable
    x = 120
    y = -456
    
    # 创建 对象实例 | Create object instance
    s = Solution()
    
    # 调用方法 输出打印结果 | Call method Output print result
    print("x =>", s.reverse(x))
    print("y =>", s.reverse(y))


````
